```python
import os
import pandas as pd
# Set the working directory
path = "/Users/venkiiyer/Documents/UMD Courses"
os.chdir(path)




def standardize_text_fields(file_path):
     """
     Reads a CSV file, converts all string columns to lowercase, and returns the cleaned DataFrame.
    
     Parameters:
     - file_path (str): The path to the CSV file.
    
     Returns:
     - pd.DataFrame: The DataFrame with standardized text fields.
     """
     # Read the CSV file
     df = pd.read_csv(file_path)
    
     # Loop through all columns
     for col in df.columns:
         # If the column is of object type (i.e., a string column), convert its values to lowercase
         if df[col].dtype == 'object':
             df[col] = df[col].str.lower()
    
     return df

standardize_text_fields("complaints.csv")



```

```python
import pandas as pd

def standardize_and_strip_text_fields(file_path):
     """
     Reads a CSV file, converts all string columns to lowercase,
     strips leading and trailing whitespace from each value, and returns the cleaned DataFrame.
    
     Parameters:
     - file_path (str): The path to the CSV file.
    
     Returns:
     - pd.DataFrame: The DataFrame with standardized and stripped text fields.
     """
     # Read the CSV file
     df = pd.read_csv(file_path)
    
     # Loop through all columns
     for col in df.columns:
         # If the column is of object type (i.e., a string column), apply cleaning operations
         if df[col].dtype == 'object':
             # Convert to lowercase and strip whitespace
             df[col] = df[col].str.lower().str.strip()
    
     return df

standardize_and_strip_text_fields("complaints.csv")
 
```

```python
import pandas as pd

def remove_tags_column_from_csv(csv_file_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_file_path)
    
    # Check if 'Tags' column exists in the DataFrame
    if 'Tags' in df.columns:
        # Drop the 'Tags' column
        df = df.drop(columns=['Tags'])
    
    # Return the remaining columns as a list
    return df.columns.tolist()

remove_tags_column_from_csv("complaints.csv")
```
